Using the FFT (Fast Fourier Transform) compute the magnitude and phase spectrum of the two images.

```python
import numpy as np
import matplotlib.pyplot as plt

# Scaling function
def scale(img):
    return ((img - img.min()) / (img.max() - img.min()) * 255)

# Load the images
x1 = np.float64(plt.imread('barbara.png'))
x2 = np.float64(plt.imread('cells.png'))

# Rescale the images in the range 0-255
x1 = scale(x1)
x2 = scale(x2)

# Compute FFT of both images
fft_x1 = np.fft.fft2(x1)
fft_x2 = np.fft.fft2(x2)

# Shift the zero-frequency component to the center
fft_x1_shifted = np.fft.fftshift(fft_x1)
fft_x2_shifted = np.fft.fftshift(fft_x2)

# Compute magnitude and phase
magnitude_x1 = np.abs(fft_x1_shifted)
magnitude_x2 = np.abs(fft_x2_shifted)
phase_x1 = np.angle(fft_x1_shifted)
phase_x2 = np.angle(fft_x2_shifted)

# Plot the magnitude and phase spectrum
plt.figure(figsize=(12, 8), dpi=80)

# Barbara image magnitude and phase
plt.subplot(221)
plt.imshow(np.log(1 + magnitude_x1), cmap='gray')
plt.title('Barbara - Magnitude Spectrum')
plt.subplot(222)
plt.imshow(phase_x1, cmap='gray')
plt.title('Barbara - Phase Spectrum')

# Cells image magnitude and phase
plt.subplot(223)
plt.imshow(np.log(1 + magnitude_x2), cmap='gray')
plt.title('Cells - Magnitude Spectrum')
plt.subplot(224)
plt.imshow(phase_x2, cmap='gray')
plt.title('Cells - Phase Spectrum')

plt.tight_layout()
plt.show()
```

Then combine the magnitude spectrum of one image with the phase spectrum of the other and vice versa. What do you observe?

```python
# Combine magnitude of x1 with phase of x2
combined_1 = np.fft.ifft2(np.fft.ifftshift(magnitude_x1 * np.exp(1j * phase_x2))).real

# Combine magnitude of x2 with phase of x1
combined_2 = np.fft.ifft2(np.fft.ifftshift(magnitude_x2 * np.exp(1j * phase_x1))).real

# Plot the reconstructed images
plt.figure(figsize=(12, 8), dpi=80)

# Original Barbara and Cells images
plt.subplot(231)
plt.imshow(x1, cmap='gray')
plt.title('Original - Barbara')
plt.subplot(232)
plt.imshow(x2, cmap='gray')
plt.title('Original - Cells')

# Combined images
plt.subplot(234)
plt.imshow(combined_1, cmap='gray')
plt.title('Magnitude of Barbara + Phase of Cells')
plt.subplot(235)
plt.imshow(combined_2, cmap='gray')
plt.title('Magnitude of Cells + Phase of Barbara')

plt.tight_layout()
plt.show()
```

What You'll Observe:
Magnitude Dominance: The visual structure of the combined image will closely resemble the image whose magnitude spectrum was used. This is because the magnitude spectrum carries the bulk of the intensity information (e.g., edges and textures).

Phase Influence: While the general structure comes from the magnitude spectrum, the phase spectrum determines the fine details, alignment, and location of features within the image. Using the phase spectrum of another image introduces distortions or shifts in these details.

Interpretation: The experiment demonstrates that the phase spectrum plays a crucial role in determining spatial structure and alignment, while the magnitude spectrum dictates the overall intensity distribution.

Add noise, $\eta$, to the "Barbara" image where

$$\eta \sim \mathcal{N}(0, \sigma^2 I)$$

Plot the original and the noisy image.

```python
# Define noise level (standard deviation)
sigma = 20  # Adjust this value to control the noise level

# Generate Gaussian noise
noise = sigma * np.random.randn(*x1.shape)

# Add noise to the "Barbara" image
x1_noisy = x1 + noise

# Clip pixel values to be in the range 0-255
x1_noisy = np.clip(x1_noisy, 0, 255)

# Plot the original and noisy images
plt.figure(figsize=(10, 5), dpi=80)

plt.subplot(121)
plt.imshow(x1, cmap='gray')
plt.title('Original - Barbara')

plt.subplot(122)
plt.imshow(x1_noisy, cmap='gray')
plt.title(f'Noisy - Barbara (={sigma})')

plt.tight_layout()
plt.show()
```

To quantify the strength of the original image $x_o$ in comparison to its noisy counterpart $x_n$, we use the Signal-to-Noise Ratio (SNR) measure, defined as:

$$ \text{SNR}(x_o,x_n) = 20\log_{10} {\frac{ \|x_o\| }{ \|x_o-x_n\| } } $$ 

It is a quantity expressed in decibels (dB). The higher the SNR, the better the signal quality.

Compute the SNR between "Barbara" and its noisy version, for different noise levels (different $\sigma$). What do you observe?

```python
def compute_snr(x_original, x_noisy):
    """Compute Signal-to-Noise Ratio (SNR) in decibels."""
    numerator = np.linalg.norm(x_original)
    denominator = np.linalg.norm(x_original - x_noisy)
    return 20 * np.log10(numerator / denominator)

# Define different noise levels
sigma_values = [5, 10, 20, 30, 50]

# Store SNR results
snr_values = []

# Compute SNR for each noise level
plt.figure(figsize=(12, 6), dpi=80)
for i, sigma in enumerate(sigma_values):
    # Generate noisy image
    noise = sigma * np.random.randn(*x1.shape)
    x1_noisy = x1 + noise
    x1_noisy = np.clip(x1_noisy, 0, 255)

    # Compute SNR
    snr = compute_snr(x1, x1_noisy)
    snr_values.append(snr)

    # Plot the noisy image
    plt.subplot(1, len(sigma_values), i + 1)
    plt.imshow(x1_noisy, cmap='gray')
    plt.title(f'={sigma}, SNR={snr:.2f} dB')

plt.tight_layout()
plt.show()

# Display SNR values
for sigma, snr in zip(sigma_values, snr_values):
    print(f" = {sigma}, SNR = {snr:.2f} dB")
```

Observations:
Noise Level vs. SNR:

As the noise level (

) increases, the SNR decreases, indicating a lower signal quality.
A high 

 results in a noisier image, reducing the difference between signal strength and noise strength.
Interpretation of SNR:

High SNR: Indicates that the signal (original image) dominates over the noise, meaning better quality.
Low SNR: Indicates that the noise is significant, degrading the image quality.

Change the image brightness of the image and compute the SNR. What do you observe?

```python
# Function to adjust brightness
def adjust_brightness(img, factor):
    """Scale image brightness by a factor."""
    return np.clip(img * factor, 0, 255)

# Define brightness scaling factors
brightness_factors = [0.5, 1.0, 1.5, 2.0, 3.0]

# Noise level for Poisson noise
sigma = 20

# Store SNR results for different brightness levels
snr_brightness = []

plt.figure(figsize=(15, 8), dpi=80)

for i, factor in enumerate(brightness_factors):
    # Adjust brightness
    x1_bright = adjust_brightness(x1, factor)

    # Add noise (Gaussian or Poisson)
    noise = sigma * np.random.randn(*x1.shape)
    x1_bright_noisy = x1_bright + noise
    x1_bright_noisy = np.clip(x1_bright_noisy, 0, 255)

    # Compute SNR
    snr = compute_snr(x1_bright, x1_bright_noisy)
    snr_brightness.append(snr)

    # Plot the noisy image
    plt.subplot(2, len(brightness_factors), i + 1)
    plt.imshow(x1_bright, cmap='gray')
    plt.title(f'Bright Factor={factor}')

    plt.subplot(2, len(brightness_factors), i + len(brightness_factors) + 1)
    plt.imshow(x1_bright_noisy, cmap='gray')
    plt.title(f'Noisy, SNR={snr:.2f} dB')

plt.tight_layout()
plt.show()

# Display SNR values
for factor, snr in zip(brightness_factors, snr_brightness):
    print(f"Brightness Factor = {factor}, SNR = {snr:.2f} dB")
```

Observations:

As the brightness of the image increases (scaling factor > 1), the SNR improves. This is because the signal (original image strength) becomes larger relative to the noise.
For dimmer images (scaling factor < 1), the SNR decreases since the signal's strength diminishes compared to the noise.
Effect of Noise:

In brighter images, the noise becomes less noticeable due to the increased pixel intensities.
In dimmer images, noise becomes more prominent and affects the visual quality significantly.
Quantitative SNR:

SNR increases logarithmically with brightness. Doubling the brightness factor roughly translates to an improvement in SNR by approximately 6 dB, as per the formula.

## 1.3 Image Blurring

We consider an image model where the observed image $y$ is a result of a degradation of an image $x$ by a linear transformation $f$, i.e.

$$
mat(y)=f(mat(x))
$$

where $mat(\cdot)$ is a notation used in this lab to underline that $f(mat(x))$ is an operation on a matrix $x$. 

Here $x$ and $y$ are images of size $512 \times 512$, then  $x\in \mathbb{R}^{512\times512}$ and  $y\in \mathbb{R}^{512\times512}$.

A common degradation is the blurring, achieved by convolving the image $x$ with a low-pass Gaussian filter kernel $h$. The blurred image $y$ is given by:

$$
mat(y)=mat(h)*mat(x) 
$$.

A common notation  is the matrix/vector notation:

$$
y= A x
$$

where $A$  is the matrix that does the convolution of the image $x$ by the Gaussian filter $h$. Here, $y$ and $x$ is the vectorized form of the images $mat(y)$ and $mat(x)$ respectiverly, in the lexical order. Then, $y\in \mathbb{R}^{N\times 1}$ and $x\in \mathbb{R}^{N\times 1}$ where $N$ is the number of pixels. It is important, as we will observe in the lab, that even though we use the notation for $A$, we do not actually construct the matrix.

### Step-by-Step Analysis:

#### 1. **Number of Pixels (\(N\))**:
The image \(x\) is of size \(512 \times 512\). Therefore, the total number of pixels is:
\[
N = 512 \times 512 = 262,144 \quad (\text{pixels})
\]

#### 2. **Size of Matrix \(A\)**:
Matrix \(A\) represents the convolution operation between the Gaussian kernel \(h\) and the image \(x\). In vectorized form:
- \(x \in \mathbb{R}^{N \times 1}\)
- \(y \in \mathbb{R}^{N \times 1}\)
- \(A\) must transform \(x\) into \(y\), so \(A\) is of size:
\[
A \in \mathbb{R}^{N \times N} = \mathbb{R}^{262,144 \times 262,144}
\]

#### 3. **Memory Size of Matrix \(A\)**:
Each element in \(A\) is a floating-point number that uses 8 bytes of memory. The total number of elements in \(A\) is:
\[
\text{Total elements in } A = N \times N = 262,144^2 = 68,719,476,736

The total memory required to store \(A\) is:
\[

Convert this to gigabytes (1 GB = \(10^9\) bytes):
\[
\text{Memory (GB)} = \frac{549,755,813,888}{10^9} = 549.76 \, \text{GB}

As you just observed, it is not always wise to create the full matrix $A$. So we will start by creating the function $f(mat(x))=mat(h)*mat(x)$, where $*$ denotes the convolution. Since convolution in time corresponds to multiplication in the frequency domain, we will do the convolution in the Fourier domain.

Compute the low pass Gaussian kernel with $\sigma=5$ pixels.  

The steps to create the convolution kernel is: 
1. Create a meshgrid of same size of the image (512x512). The center of the meshgrid could be 0. 
2. Apply the the Gaussian function onto the grid, and normalize the response such that the sum of all elements in the reponse =1. 
3. Apply the Fast Fourier Transformation. NOTE: If the center of the grid was defined as 0, a fftshift has to be done before applying the FFT2. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create the meshgrid
image_size = 512
sigma = 5  # Standard deviation of the Gaussian

# Define x and y coordinates centered at 0
x = np.linspace(-image_size // 2, image_size // 2 - 1, image_size)
y = np.linspace(-image_size // 2, image_size // 2 - 1, image_size)
X, Y = np.meshgrid(x, y)

# Step 2: Apply the Gaussian function
gaussian_kernel = np.exp(-(X**2 + Y**2) / (2 * sigma**2))

# Normalize the kernel so that the sum of all elements is 1
gaussian_kernel /= gaussian_kernel.sum()

# Step 3: Apply FFT
# Perform fftshift to center the frequency components, then apply fft2
gaussian_kernel_fft = np.fft.fft2(np.fft.fftshift(gaussian_kernel))

# Visualization
plt.figure(figsize=(12, 5), dpi=80)

# Plot the Gaussian kernel in the spatial domain
plt.subplot(121)
plt.imshow(gaussian_kernel, cmap='gray', extent=(-image_size // 2, image_size // 2, -image_size // 2, image_size // 2))
plt.title('Gaussian Kernel (Spatial Domain)')
plt.colorbar()

# Plot the magnitude of the Fourier Transform of the kernel
plt.subplot(122)
plt.imshow(np.log1p(np.abs(gaussian_kernel_fft)), cmap='gray')  # Use log scale for better visualization
plt.title('FFT of Gaussian Kernel (Frequency Domain)')
plt.colorbar()

plt.tight_layout()
plt.show()
```

# 2. Introduction to Inverse Problems

A more realistic model writes
$$
mat(y)=f(mat(x)) + mat(\eta) \qquad (1)
$$
where $mat(\cdot)$ is a notation used in this lab to underline that $f(mat(x))$ is an operation on a matrix $x$. 

A more common notation to write (1) is: 
$$
y= Ax + \eta
$$
where $A$ is the linear transformation and $x$, $y$ and $\eta$ are viewed as vectors. However, when working in image processing the matrix $A$ will rearly be constructed, and we will not reshape the image $x$ as a vector.  It is important, as we have observed in the lab, that even though we use the notations $A$, we do not actually construct the matrix. 

In this lab, we considerer that the noise, $\eta$ follows a multidimensional normal law of covarience $\sigma^2 I$ and of mean the zero vector. So, we can write: $\eta \sim \mathcal{N}(0, \sigma^2 I)$. 
  
The probability density function $p_\eta(\eta)$ is written as: 
\begin{equation}
 p_\eta(\eta) = \frac{1}{(2\pi\sigma^2)^{\frac{N}{2}}} \exp \left(-\frac{\|\eta\|_2^2}{2\sigma^2}\right)
\end{equation}
 where $N$ is the number of pixels and $\|\eta\|_p$ is the $l^p$-norm defined as: 
\begin{equation}
  \|\eta\|_p = \left(\sum_{i=1}^N |\eta|^p\right)^{\frac{1}{p}}
\end{equation}
 
 We want to find the unknown image $x$ from the observation $y$. Therefore, we use the  maximum likelihood estimation which maximizes the likelihood $L(y,x)$ with respect to the unknown image $x$. This likelihood is equal to the conditional probability of $y$ knowing $x$, denoted $p_{y|x}(y|x)$. The probability is calculated from the image model, supposing that the $\eta$ is white Gaussian noise. 

More precisely, the likelihood $L(y,x)$ is given by:
\begin{equation}
 L(y,x) = p_{y|x}(y|x) = p_n(n=Ax - y) = \frac{1}{(2\pi\sigma^2)^{\frac{N}{2}}} \textrm{exp}\left(-\frac{\|Ax-y\|_2^2}{2\sigma^2}\right)
\end{equation}

We search for an estimation $\hat{x}$ of the real image $x$ by maximizing the ligelihood  $L(y,x)$ as follows: 
\begin{equation}
 \hat{x} = \arg\max_x L(y,x)
\end{equation}

In order to avoid the difficulties related to the exponential, we often maximize the logarithm of the likelihood (which does not change the maximum argument since the logarithm is strictly increasing):
\begin{equation}
 \hat{x} = \arg\max_x \, \ln(L(y,x)) = \arg\max_x \, \left(- \ln \left(2\pi\sigma^2 \right)^{\frac{N}{2}} -\frac{1}{2\sigma^2}\|Ax-y\|_2^2\right)
\end{equation}

The term $- \ln \left( 2\pi\sigma^2 \right)^{\frac{N}{2}}$ is a constant with respect to $x$, and thus does not intervene in the estimation of $\arg\max$. Therefore, we can write:
\begin{equation}
 \hat{x} = \arg\max_x \, \left( -\frac{1}{2\sigma^2}\|Ax-y\|_2^2\right)
\end{equation}

The last step removes the proptionality coefficent $\frac{1}{2\sigma^2}$ and the negative sign by using the fact that  $\arg\max_x -f(x) = \arg\min_x f(x)$. So, finally: 

\begin{equation*}
 \hat{x} = \arg\min_x \, \|Ax-y\|_2^2 
\end{equation*}


Denote $J(x)=\|Ax-y\|_2^2 $. This term is in fact the Least Square term associated to the system $y = Ax$. The gradient of $J$ is $\nabla J(x)= 2A^*(Ax-y)$, where $A^*$ is the adjoint operator of A.

#### The importance of Matrix Condition number
The matrix condition number can be calculated, for an $A$ invertible, as 
$$cond(A) = \|A\|\|A^{-1}\|$$
Where $\|A\|= \sup_{x \neq 0} \frac{\|Ax\|}{\|x\|}$.
It can also be calculated  
$$cond(A) = \frac{\lambda_{max}}{\lambda_{min}}$$

If we have $y=Ax$, then $\|y\|\leq \|A\|\|x\|$.
If we have a small pertubation, then 
$y+\delta y = A(x+ \delta x)$, and $\delta y= A \delta x$. Which yields  $\|\delta x\|\leq \|A^{-1}\|\|\delta y\|$.
And thus, 

$$\frac{\|\delta x\|}{\|x\|}\leq\|A\| \|A^{-1}\|\frac{\|\delta y\|}{\|y\|}$$
or, equivalently
$$\frac{\|\delta x\|}{\|x\|}\leq cond(A) \frac{\|\delta y\|}{\|y\|}$$

Why is the Matrix Condition number important ? 
What is the condition number for our problem (remember that the eigen values of $A$ are given by (equal to) the coefficients of the Fourier transform of h)?  Is the problem ill-posed?
